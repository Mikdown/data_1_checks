```python
import pandas as pd
import requests
import json
url_local_endpoints = "https://api.weather.gov/points/38.199371,-85.653526"
url_local_forecast = "https://api.weather.gov/gridpoints/LMK/52,75/forecast"
```

```python
headers = {'Accept': 'application/json'}
data_endpoints = requests.get(url_local_endpoints, headers=headers)
response = data_endpoints.json()
with open('assets/data_endpoints.json', 'w') as json_file:
    json.dump(response, json_file)


```

```python
headers = {'Accept': 'application/json'}
local_forecast = requests.get(url_local_forecast, headers=headers)
response1 = local_forecast.json()
with open('assets/local_forecast.json', 'w') as json_file:
    json.dump(response1, json_file)
```

```python
response1_df = pd.json_normalize(response1)

response1_df
```

```python
nested_response1_df = pd.json_normalize(response1, ["properties", "periods"])

nested_response1_df
```
