```python
import pandas as pd
```

### The following code cell reads a "dirty" .csv file as the source to create a dataframe.

```python
ufo_df = pd.read_csv("assets/ufo_report.csv")
```

### The folowing code cell prints the first 10 rows of the dataframe before "cleaning" for reference.

```python
ufo_df.head(10)
```

### The following code cell meets the data check requirement by fixing 6 issues with the dataframe and then saving the "cleaned" dataframe to a new .csv file. It then prints the first 10 rows for verification.

```python
ufo_df.columns = [x.capitalize() for x in ufo_df.columns] #Capitalize all column headers.
ufo_df.columns = ufo_df.columns.str.replace('_', '-') #Replaces underscore with hyphen in column headers.
ufo_df.rename({'Text': 'Description'}, axis=1, inplace=True) #Renames "Text" column to "Description".
ufo_df.drop(["Unnamed: 0"], axis=1, inplace=True) #Removes the unused index column.
ufo_df.Duration = ufo_df.Duration.fillna('Unknown') #Replaces NaN with "Unknown" in "Duration" column.
ufo_df.State = ufo_df.State.fillna('N/A') #Replaces NaN with "N/A" in "State" column.
ufo_df.to_csv("assets/ufo_report_cleaned.csv")
ufo_df.head(10)
```
